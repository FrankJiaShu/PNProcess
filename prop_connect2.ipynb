```python
import pandas as pd
```

### 读取压缩后的数据

```python
%%time
select = pd.read_csv('./result/words_selected_result.csv')
select
```

### 读取属性表

```python
%%time
props = pd.read_excel('./dataset/prop2.xlsx')
props
```

### 制作字典

```python
def create_dict(props):
    prop_dict = {}
    for i in range(len(props)):
        id = props.iloc[i, 0]
        fields = props.iloc[i, 2:5].values # 取三列属性abc
        prop_dict[id] = fields
    return prop_dict
```

```python
%%time
prop_dict = create_dict(props)
prop_dict
```

### 加权求和处理

```python
def sum_fields(select, prop_dict):
    res = []
    for i in range(len(select)):
        row = select.iloc[i, :].values
        # row[0]为单词 其他为id号
        sum = [0, 0, 0]
        for j in range(1, len(row)):
            # 去取属性
            if row[j] > 0.0:
                values = prop_dict.get(j)
                if values is not None:
                    sum[0] += row[j] * values[0]
                    sum[1] += row[j] * values[1]
                    sum[2] += row[j] * values[2]
        res.append([row[0], sum[0], sum[1], sum[2]]) # 保存单词-属性a求和-属性b求和-属性c求和
    return pd.DataFrame(res, columns=['word', 'a', 'b', 'c'])
```

```python
%%time
res = sum_fields(select, prop_dict)
res
```

### 生成结果文件

```python
# 传入df和文件名生成csv文件
def save_file(df, file_name):
    df.to_csv('./result/' + file_name + 'result.csv', index=False)
```

```python
%%time
save_file(res, 'prop_connect2_')
```
