## 属性连接1-技术领域去重

```python
import pandas as pd
```

### 读取压缩后的数据

```python
%%time
select = pd.read_csv('./result/words_selected_result.csv')
select
```

### 读取属性表

```python
%%time
props = pd.read_excel('./dataset/prop1.xlsx')
props
```

### 制作字典

```python
def create_dict(props):
    prop_dict = {}
    for i in range(len(props)):
        id = props.iloc[i, 0]
        field = props.iloc[i, 2]
        if not pd.isnull(field):
            fields = str(field).strip(' ').split(',')
            prop_dict[id] = fields
    return prop_dict
```

```python
%%time
prop_dict = create_dict(props)
prop_dict
```

### 映射处理技术领域

```python
def mapping_fields(select, prop_dict):
    res = []
    for i in range(len(select)):
        fields = set()
        row = select.iloc[i, :].values
        # row[0]为单词 其他为id号
        for j in range(1, len(row)):
            if (row[j] > 0.0): # 如果列大于0就去取技术领域
                values = prop_dict.get(j)
                if values is not None:
                    for k in values:
                        fields.add(k)
        res.append([row[0], list(fields), len(fields)]) # 保存单词-技术领域-数量
        
    return pd.DataFrame(res, columns=['word', 'fields', 'sum'])
```

```python
%%time
res = mapping_fields(select, prop_dict)
res
```

### 生成结果文件

```python
# 传入df和文件名生成csv文件
def save_file(df, file_name):
    df.to_csv('./result/' + file_name + 'result.csv', index=False)
```

```python
%%time
save_file(res, 'prop_connect1_')
```
